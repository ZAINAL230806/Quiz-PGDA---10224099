```python
import random as rnd
import matplotlib.pyplot as plt

def mean_distance_2D(Nrep, Nstep=101):
    """Menghitung rata-rata jarak (mean distance) dalam random walk 2D."""
    mean_r = [0.0] * Nstep

    for j in range(int(Nrep)):
        x, y = 0.0, 0.0
        for i in range(Nstep):
            direction = rnd.randint(0, 3)
            if direction == 0:
                x += 1   # Timur
            elif direction == 1:
                x -= 1   # Barat
            elif direction == 2:
                y += 1   # Utara
            else:
                y -= 1   # Selatan

            r = (x**2 + y**2)**0.5
            mean_r[i] += r

    # Rata-rata jarak untuk tiap langkah
    for i in range(Nstep):
        mean_r[i] /= Nrep

    return mean_r


# ===============================
# Hitung untuk berbagai Nrep
# ===============================
sx_050 = mean_distance_2D(50)
sx_100 = mean_distance_2D(100)
sx_200 = mean_distance_2D(200)
sx_400 = mean_distance_2D(400)
sx_800 = mean_distance_2D(800)

# ===============================
# Plot hasil simulasi
# ===============================
plt.figure(figsize=(8,5))
plt.plot(sx_050, label='Nrep = 50')
plt.plot(sx_100, label='Nrep = 100')
plt.plot(sx_200, label='Nrep = 200')
plt.plot(sx_400, label='Nrep = 400')
plt.plot(sx_800, label='Nrep = 800')
plt.xlabel('Steps')
plt.ylabel('Mean Distance ⟨r⟩')
plt.title('2D Random Walk - Influence of Nrep on Mean Distance')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random as rnd
import matplotlib.pyplot as plt

# Fungsi untuk menghitung mean squared distance
def mean_squared_distance(Nrep, Nstep=101):
    mean_r2 = [0.0] * Nstep
    for j in range(Nrep):
        x, y = 0.0, 0.0
        for i in range(Nstep):
            direction = rnd.randint(0, 3)
            if direction == 0:
                x += 1
            elif direction == 1:
                x -= 1
            elif direction == 2:
                y += 1
            else:
                y -= 1
            r2 = x**2 + y**2
            mean_r2[i] += r2
    for i in range(Nstep):
        mean_r2[i] /= Nrep
    return mean_r2


# Hitung mean squared distance untuk berbagai Nrep
sx2_050 = mean_squared_distance(50)
sx2_100 = mean_squared_distance(100)
sx2_200 = mean_squared_distance(200)
sx2_400 = mean_squared_distance(400)
sx2_800 = mean_squared_distance(800)

# Plot hasilnya
plt.figure(figsize=(8,5))
plt.xlabel('steps')
plt.ylabel('mean squared distance ⟨r²⟩')
plt.grid(True)
plt.plot(sx2_050, label='Nrep=50')
plt.plot(sx2_100, label='Nrep=100')
plt.plot(sx2_200, label='Nrep=200')
plt.plot(sx2_400, label='Nrep=400')
plt.plot(sx2_800, label='Nrep=800')
plt.legend()
plt.title('2D Random Walk — Influence of Nrep on Mean Squared Distance')
plt.show()
```
