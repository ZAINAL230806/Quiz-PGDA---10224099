```python
import random as rnd
import matplotlib.pyplot as plt

def mean_distance_1D(Nrep, Nstep=21):
    """Menghitung rata-rata jarak (mean distance) untuk random walk 1D."""
    sx = [0.0] * Nstep

    for j in range(int(Nrep)):
        x = [0.0] * Nstep
        pos = 0
        for i in range(Nstep):
            step = 2 * rnd.randint(0, 1) - 1  # langkah: 1
            pos += step
            x[i] = abs(pos)  # jarak dari titik awal
        for i in range(Nstep):
            sx[i] += x[i]

    # ambil rata-ratanya
    for i in range(Nstep):
        sx[i] /= Nrep

    return sx


# ============================
# Hitung untuk berbagai Nrep
# ============================
sx_1E2 = mean_distance_1D(1e2)
sx_1E3 = mean_distance_1D(1e3)
sx_1E4 = mean_distance_1D(1e4)
sx_1E5 = mean_distance_1D(1e5)
# (catatan: 1E6 butuh waktu lama, jadi hati-hati kalau dijalankan)
# sx_1E6 = mean_distance_1D(1e6)

# ============================
# Plot hasil simulasi
# ============================
plt.figure(figsize=(8,5))
plt.plot(sx_1E2, label='Nrep = 1E2')
plt.plot(sx_1E3, label='Nrep = 1E3')
plt.plot(sx_1E4, label='Nrep = 1E4')
plt.plot(sx_1E5, label='Nrep = 1E5')
# plt.plot(sx_1E6, label='Nrep = 1E6')  # opsional
plt.xlabel('Steps')
plt.ylabel('Mean Distance')
plt.title('1D Random Walk - Influence of Nrep on Mean Distance')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import random as rnd
import matplotlib.pyplot as plt

def mean_squared_distance_1D(Nrep, Nstep=121):
    """Menghitung rata-rata kuadrat jarak (mean squared distance) dalam random walk 1D."""
    sx2 = [0.0] * Nstep

    for j in range(int(Nrep)):
        pos = 0
        for i in range(Nstep):
            step = 2 * rnd.randint(0, 1) - 1  # langkah acak 1
            pos += step
            sx2[i] += pos ** 2

    # Hitung rata-rata
    for i in range(Nstep):
        sx2[i] /= Nrep

    return sx2


# ============================
# Hitung MSD untuk berbagai Nrep
# ============================
sx_025 = mean_squared_distance_1D(25)
sx_050 = mean_squared_distance_1D(50)
sx_100 = mean_squared_distance_1D(100)
sx_200 = mean_squared_distance_1D(200)
sx_400 = mean_squared_distance_1D(400)

# ============================
# Plot hasil simulasi
# ============================
plt.figure(figsize=(8,5))
plt.plot(sx_025, label='Nrep = 25')
plt.plot(sx_050, label='Nrep = 50')
plt.plot(sx_100, label='Nrep = 100')
plt.plot(sx_200, label='Nrep = 200')
plt.plot(sx_400, label='Nrep = 400')
plt.xlabel('Steps')
plt.ylabel('Mean Squared Distance x')
plt.title('1D Random Walk - Influence of Nrep on Mean Squared Distance')
plt.grid(True)
plt.legend()
plt.show()
```
